```python
#spectral centroid -- centre of mass -- weighted mean of the frequencies present in the sound
import sklearn
spectral_centroids = librosa.feature.spectral_centroid(data1, sr=sr)[0]
spectral_centroids.shape
# Computing the time variable for visualization
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)
# Normalising the spectral centroid for visualisation
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)
#Plotting the Spectral Centroid along the waveform
librosa.display.waveplot(data1, sr=sr, alpha=0.4)
plt.plot(t, normalize(spectral_centroids), color='r')
```

```python
mfccs = librosa.feature.mfcc(data1, sr=sr)
print(mfccs.shape)
print(mfccs)
#Displaying  the MFCCs:
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
```

```python
import numpy as np
```

```python
all_features=[]
for genre in genres:
    for file in os.listdir(f'mini-genres/mini-genres/{genre}'):
        songname = f'mini-genres/mini-genres/{genre}/{file}'
        print(songname)
        y,sr=librosa.load(songname,mono=True,duration=30)
        chroma= librosa.feature.chroma_stft(y,sr=sr)
        spectral = librosa.feature.spectral_centroid(y, sr=sr)
        mfcc = librosa.feature.mfcc(y , sr=sr)
        feature_list = [np.mean(chroma),np.mean(spectral)]
        for feat in mfcc[:15]:
            feature_list.append(np.mean(feat))
        feature_list.append(genre)
        all_features.append(feature_list)
```

```python
all_features
```

```python
df = pd.DataFrame(all_features,columns=df.columns)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder , StandardScaler
```

```python
scalar = StandardScaler()
```

```python
X = scalar.fit_transform(np.array(df.iloc[:,:-1],dtype=float))
X.shape
```

```python
y = df['genre']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.5, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
from sklearn import metrics
print('Accuracy :',metrics.accuracy_score(y_test,y_pred))
```
