```python
import pandas as pd
```

```python
dff=pd.read_csv("kc_house_data.csv")
dff.head()
```

```python
dff.isnull().sum()
```

```python
dff['sqft_above'].fillna(dff['sqft_above'].median(),inplace=True)
dff.isnull().sum()
```

```python
dff=dff.drop('id',axis=1)
dff=dff.drop('zipcode',axis=1)
dff=dff.drop('lat',axis=1)
dff=dff.drop('long',axis=1)
dff=dff.drop('date',axis=1)
```

```python
dff.head()
```

```python
x=dff.filter(['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','grade','sqft_above','sqft_basement','yr_built','yr_renovated','sqft_living15','sqft_lot15'])
y=dff['price']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=424)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
predicted= reg.predict(X_test)
print(predicted)
```

```python
import sklearn.metrics as sm
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, predicted), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(y_test,predicted), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(y_test, predicted), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(y_test, predicted), 2)) 
print("R2 score =", round(sm.r2_score(y_test, predicted), 2))
```
